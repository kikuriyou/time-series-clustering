# 時系列データのクラスタリング(k-means)
---
- 階層クラスタリングではデータセットが大きくなると計算が遅くなるので、k-meansでも試す

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
```

## 前処理
---
元のファイルでは数字ごとの区切りに";"、小数点に","が用いられているために、これらをそれぞれ","と"."に置換する

```python
# ファイルをパースして区切り文字を置換
with open('../input/LD2011_2014.txt', 'r') as f:
    file = f.read()

file = file.replace(',', '.')
file = file.replace(';', ',')

with open('../input/LD2011_2014_parsed.txt', 'w') as f:
    f.write(file)
```

区切り文字の置換後に再度ファイルを読み込み

```python
nrows = None
df = pd.read_csv('../input/LD2011_2014_parsed.txt', nrows=nrows)
df.head(5)
```

```python
# 型をfloatに統一
for col in df.drop('Unnamed: 0', axis=1).columns:
    df[col] = df[col].astype(float)
```

```python
n_sample = 1000
np.random.seed(123)
df_use = df.drop('Unnamed: 0', axis=1).sample(n=n_sample).reset_index(drop=True)

# サンプル同士で値に違いがない(=すべてゼロ)カラムは削除
for col in df_use.columns:
    if df_use[col].nunique() == 1:
        df_use = df_use.drop(col, axis=1)

# 正規化
max_val = df_use.max().max()
df_use = df_use / max_val

# ゼロ割り回避
df_use = df_use.replace(0, 1e-10)

display(df_use.sample(n=5))
print(df_use.shape)
df_use.head(5).T.plot(); plt.show()
```

## 適切なクラスター数の探索
---
エルボー法
- クラスター数ごとに、distortionをプロットする
- 基本的にクラスター数が多くなればdistortionは小さくなるが、適切なクラスター数を超えるとその勾配は小さくなる

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

x_train = df_use.copy()

scaler = StandardScaler()
x_trn = scaler.fit_transform(x_train)

distortions = []
n_clusters_try = 10
for i in range(1, n_clusters_try+1):
    km = KMeans(n_clusters=i, init='k-means++', random_state=123)
    km.fit(x_trn)
    #y_km = km.fit_predict(x_trn)
    distortions.append(km.inertia_)

plt.plot(range(1, n_clusters_try+1), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Elbow method')
plt.show()
```

クラスター数:3 が適切と言える

シルエット図
- クラスター内外のサンプルとの平均距離から求められるシルエット係数の分布を表示する
- シルエット係数の範囲は[-1, 1]で、1に近いほど他のクラスターから離れていることを示す

```python
from sklearn.metrics import silhouette_samples
from matplotlib import cm

plt.figure(figsize=(16,20))
n_clusters_try = 10
i_plt = 0

for i in range(2, n_clusters_try+1):
    i_plt += 1
    
    # kmean実行
    km = KMeans(n_clusters=i, init='k-means++', random_state=123)
    km.fit(x_trn)
    y_km = km.fit_predict(x_trn)
    
    cluster_labels = np.unique(y_km)
    n_clusters = len(cluster_labels)
    
    # シルエット係数を算出
    silhouette_vals = silhouette_samples(x_trn, y_km, metric='euclidean')
    
    # 図化
    plt.subplot(3, 3, i_plt)
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        plt.barh(range(y_ax_lower, y_ax_upper),
                c_silhouette_vals,
                height=1.0,
                edgecolor='none',
                color=color)
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)
        
    silhouette_mean = np.mean(silhouette_vals)
    plt.axvline(silhouette_mean, color='red', linestyle="--")
    plt.yticks(yticks, cluster_labels)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.title('n_cluster: {}'.format(n_clusters))
plt.tight_layout()
plt.show
```

- 赤線：全てのシルエット係数の平均値
- 各クラスターのシルエット係数の最大値がシルエット係数の平均値より大きいことから、クラスター数:2~4程度 が適切と考えられる

```python
km = KMeans(n_clusters=3, init='k-means++', random_state=123)
km.fit(x_trn)
y_km = km.fit_predict(x_trn)

cluster_idx = y_km
print('Number of cluster:', len(np.unique(cluster_idx)))

# プレイ時間のヒートマップをクラスタリング結果の順番に表示
df_clustered = pd.concat([
    df_use,
    pd.DataFrame(cluster_idx, columns=['cluster'])
], axis=1)
df_clustered = df_clustered.sort_values(by='cluster')

plt.imshow(df_clustered, aspect='auto', interpolation='none', vmin=-0.2, vmax=0.4, cmap='Blues')
plt.colorbar(pad=0.2)
plt.gca().yaxis.tick_right()
plt.tick_params(labelbottom='off')
plt.show()
```

```python
# クラスターごとに時系列データを表示
for idx in np.unique(cluster_idx):
    plt.subplot(8, 1, idx+1)
    for i in range(len(df_use.index)):
        if cluster_idx[i]==idx:
            df_use.iloc[i, :].plot(figsize = (18, 24))
            plt.ylim(0, 1)
            #plt.xlim(20, 40)
            plt.title('Cluster:{}'.format(idx))
            plt.tight_layout()
plt.show()
```

比較的大きく増加しているところが3箇所あり、3箇所目の違いが大きい → 3つの時期を拡大して比較する

```python
# 1つ目
for idx in np.unique(cluster_idx):
    plt.subplot(8, 4, idx+1)
    for i in range(len(df_use.index)):
        if cluster_idx[i]==idx:
            df_use.iloc[i, :].plot(figsize = (18, 24))
            plt.ylim(0, 1)
            plt.xlim(180, 200)
            plt.title('Cluster:{}'.format(idx))
            plt.tight_layout()
plt.show()

# 2つ目
for idx in np.unique(cluster_idx):
    plt.subplot(8, 4, idx+1)
    for i in range(len(df_use.index)):
        if cluster_idx[i]==idx:
            df_use.iloc[i, :].plot(figsize = (18, 24))
            plt.ylim(0, 1)
            plt.xlim(270, 290)
            plt.tight_layout()
plt.show()
            
# 3つ目
for idx in np.unique(cluster_idx):
    plt.subplot(8, 4, idx+1)
    for i in range(len(df_use.index)):
        if cluster_idx[i]==idx:
            df_use.iloc[i, :].plot(figsize = (18, 24))
            plt.ylim(0, 1)
            plt.xlim(350, 370)
            plt.tight_layout()
plt.show()
```

各クラスターの特徴
- cluster:0 3つ目の増加が大きい
- cluster:1 3つ目の増加がない
- cluster:2 cluster:0より小さいが、3つ目の増加がある
